### Encode Data 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X_Independent[:, 0] = labelencoder_X.fit_transform(X_Independent[:, 0])
X_Independent[:, 1] = labelencoder_X.fit_transform(X_Independent[:, 1])
X_Independent[:, 2] = labelencoder_X.fit_transform(X_Independent[:, 2])
X_Independent[:, 3] = labelencoder_X.fit_transform(X_Independent[:, 3])
X_Independent[:, 4] = labelencoder_X.fit_transform(X_Independent[:, 4])


labelencoder_Y = LabelEncoder()
Y_Dependent = labelencoder_Y.fit_transform(Y_Dependent)
```

```python
for i in range(len(X_Independent)):
    if X_Independent[i][1] == 7:
        X_Independent[i][1] = np.nan
    if X_Independent[i][3] == 1:
        X_Independent[i][3]= np.nan
    if X_Independent[i][4] == 0:
        X_Independent[i][4]= np.nan
```

### Mengisi Missing Value dengan KNN 

```python
from fancyimpute import KNN
```

```python
X_filled_knn = KNN(k=3).fit_transform(X_Independent)
```

### Split Dataset Menjadi Train dan Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_filled_knn,Y_Dependent, test_size = 0.2, random_state = 200)
```

### Scalling Menggunakan Standar Scalar

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Klasifikasi (SVM)

```python
from sklearn.svm import SVC
```

### Hyperparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV 


param_grid = {'C': [0.1, 1, 10, 100, 1000], 
        'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
        'kernel': ['rbf']} 

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3) 


grid.fit(X_train, Y_train) 
```

```python
print(grid.best_estimator_) 
```

```python
svc = SVC(C=1000, cache_size=200, class_weight=None, coef0=0.0,
    decision_function_shape='ovr', degree=3, gamma=1, kernel='rbf', max_iter=-1,
    probability=False, random_state=None, shrinking=True, tol=0.001,
    verbose=False)
svc.fit(X_train,Y_train)
```

```python
y_pred = svc.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred))
```
